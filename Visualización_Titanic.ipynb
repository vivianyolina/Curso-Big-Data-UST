```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Cargamos la base de datos desde un archivo CSV en Colab

df = pd.read_csv("BaseDeDatosTitanic.csv")

# Mostramos las primeras filas de la base de datos para revisar su estructura
df.head()
```

##Introducción a la data

El conjunto de datos proporcionado se refiere a información sobre los pasajeros que viajaron a bordo del RMS Titanic durante su trágico viaje inaugural. Este conjunto de datos es ampliamente utilizado para explorar patrones y factores asociados con la supervivencia de los pasajeros en el Titanic.

El hundimiento del Titanic ocurrió en la madrugada del 15 de abril de 1912, tras chocar con un iceberg, resultando en una de las tragedias más notorias en la historia de la navegación. Durante el accidente, muchos pasajeros perdieron la vida, mientras que otros lograron sobrevivir. Este conjunto de datos contiene información sobre una muestra de pasajeros y presenta diversas variables que pueden ayudar a entender qué factores influyeron en las posibilidades de supervivencia.

Variables del conjunto de datos:

**Survived**: Variable categórica que indica si el pasajero sobrevivió o no al desastre. Tiene dos categorías: 0 para "No sobrevivió" y 1 para "Sobrevivió".

**Pclass**: Variable categórica que representa la clase de pasajero. Tiene tres categorías: 1 para "Primera clase", 2 para "Segunda clase" y 3 para "Tercera clase".

**Sex**: Variable categórica que indica el género del pasajero. Tiene dos categorías: "Male" para hombres y "female" para mujeres.

**Age**: Variable cuantitativa que representa la edad del pasajero en años. Esta variable proporciona información sobre la distribución de edades entre los pasajeros.

**SibSp**: Variable cuantitativa que indica el número de hermanos o cónyuges que el pasajero tenía a bordo del Titanic. Esta variable puede ayudar a entender la estructura familiar de los pasajeros.

**Parch**: Variable cuantitativa que representa el número de padres o hijos que el pasajero tenía a bordo del Titanic. Al igual que SibSp, esta variable proporciona información sobre las relaciones familiares.

**Fare**: Variable cuantitativa que muestra la tarifa pagada por el pasajero para su viaje en el Titanic. Esta variable puede reflejar la clase del pasajero y otros detalles relacionados con su viaje.

**Embarked**: Variable categórica que indica el puerto de embarque del pasajero. Tiene tres categorías: "C" para Cherbourg, "Q" para Queenstown y "S" para Southampton.

Con estas variables, podemos explorar diversas relaciones y patrones en los datos para comprender mejor qué características podrían haber influido en las posibilidades de supervivencia de los pasajeros en el trágico evento del Titanic.
