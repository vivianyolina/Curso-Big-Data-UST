# 4. Seleccione 2 variables cualitativas:

##a) Realice tablas resúmenes o de frecuencia y comente.


```python
# Selección de las dos variables cualitativas: Survived y Pclass
selected_data = data[["Survived", "Pclass"]]

# Tabla de frecuencia para la variable "Survived"
survived_freq = selected_data["Survived"].value_counts()

# Tabla de frecuencia para la variable "Pclass"
pclass_freq = selected_data["Pclass"].value_counts()

print("Tabla de frecuencia para la variable 'Survived':")
print(survived_freq)

print("\nTabla de frecuencia para la variable 'Pclass':")
print(pclass_freq)
```

Análisis de las tablas de frecuencia:

La tabla de frecuencia para la variable "Survived" muestra que hay 549 pasajeros que no sobrevivieron (Survived=0) y 340 pasajeros que sí sobrevivieron (Survived=1). Esto significa que de la muestra de 889 pasajeros, aproximadamente 549 no lograron sobrevivir al naufragio del Titanic, mientras que aproximadamente 340 sí lograron sobrevivir.

Por lo tanto, podemos calcular las proporciones o porcentajes de sobrevivientes y no sobrevivientes:

Proporción de no sobrevivientes: 549 / 889 ≈ 0.617 ≈ 61.7%
Proporción de sobrevivientes: 340 / 889 ≈ 0.383 ≈ 38.3%
Esto nos permite tener una idea de la tasa de supervivencia en la muestra, donde aproximadamente el 61.7% de los pasajeros no sobrevivió y aproximadamente el 38.3% sí sobrevivió.

La tabla de frecuencia para la variable "Pclass" muestra la cantidad de pasajeros en cada clase de pasajero. En la muestra, hay 491 pasajeros en la Tercera Clase, 214 pasajeros en la Primera Clase y 184 pasajeros en la Segunda Clase.

Esto nos permite tener una idea de la distribución de pasajeros en cada clase:

Tercera Clase: 491 pasajeros
Primera Clase: 214 pasajeros
Segunda Clase: 184 pasajeros
Podemos observar que la mayoría de los pasajeros (491 de 889) viajaban en la Tercera Clase, mientras que la Primera Clase tenía menos pasajeros (214 de 889) y la Segunda Clase tenía una cantidad intermedia de pasajeros (184 de 889).

##b) Gráfico de barras y comente hallazgos.


```python
import matplotlib.pyplot as plt

# Gráfico de barras para la variable "Survived"
plt.figure(figsize=(6, 4))
plt.bar(survived_freq.index, survived_freq.values)
plt.title("Gráfico de Barras - Sobrevivientes")
plt.xlabel("Sobrevivido (Survived)")
plt.ylabel("Cantidad de Pasajeros")
plt.xticks([0, 1], labels=["No Sobrevivieron", "Sobrevivieron"])
plt.show()

# Gráfico de barras para la variable "Pclass"
plt.figure(figsize=(6, 4))
plt.bar(pclass_freq.index, pclass_freq.values)
plt.title("Gráfico de Barras - Clase de Pasajero")
plt.xlabel("Clase de Pasajero (Pclass)")
plt.ylabel("Cantidad de Pasajeros")
plt.xticks([1, 2, 3], labels=["Primera Clase", "Segunda Clase", "Tercera Clase"])
plt.show()
```

**Gráfico de Barras - Sobrevivientes:**
En este gráfico, podemos ver dos barras, una correspondiente a los pasajeros que no sobrevivieron (Survived=0) y otra a los pasajeros que sí sobrevivieron (Survived=1). La altura de cada barra representa la cantidad de pasajeros en cada categoría.

**Interpretación:**

La barra "No Sobrevivieron" (Survived=0) es más alta que la barra "Sobrevivieron" (Survived=1), lo que indica que hubo más pasajeros que no sobrevivieron al naufragio del Titanic que pasajeros que sobrevivieron.
La diferencia en la altura de las barras muestra que el número de no sobrevivientes supera al número de sobrevivientes, lo que resalta la gravedad de la tragedia y el alto costo humano del desastre.



**Gráfico de Barras - Clase de Pasajero:**
En este gráfico, tenemos tres barras, una para cada clase de pasajero: Primera Clase (Pclass=1), Segunda Clase (Pclass=2) y Tercera Clase (Pclass=3). La altura de cada barra representa la cantidad de pasajeros en cada clase.

**Interpretación:**

La barra "Tercera Clase" (Pclass=3) es la más alta, lo que indica que la mayoría de los pasajeros a bordo del Titanic pertenecían a la Tercera Clase.
La barra "Primera Clase" (Pclass=1) es la siguiente más alta, lo que indica que había una cantidad significativamente menor de pasajeros viajando en la Primera Clase en comparación con la Tercera Clase.
La barra "Segunda Clase" (Pclass=2) es la más corta de las tres, lo que sugiere que había menos pasajeros en la Segunda Clase en comparación con la Tercera y Primera Clase.


En resumen, los gráficos de barras nos proporcionan una representación visual clara de la distribución y frecuencia de las variables cualitativas "Survived" (Sobrevivido) y "Pclass" (Clase de pasajero). Nos permiten identificar rápidamente las diferencias en la cantidad de pasajeros que sobrevivieron o no sobrevivieron, así como la distribución de pasajeros en cada clase de pasajero a bordo del Titanic.
